```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn 
%matplotlib inline
```

```python
x = np.linspace(-6, 6, num = 1000)
plt.figure(figsize = (12, 8))
plt.plot(x, 1 / (1 + np.exp(-x))) # Sigmoid function
plt.xlabel('x')
plt.ylabel('y')
plt.title('Sigmoid function')
```
